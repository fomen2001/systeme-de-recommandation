SESSION 1 : Découverte et Exploration des Données
1. Importation des librairies et chargement des données
But : Charger les fichiers bruts (u.data, u.item, u.user) dans des DataFrames Pandas pour pouvoir les manipuler. Nous devons gérer les séparateurs spécifiques (| et tabulations) et l'encodage.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics.pairwise import cosine_similarity

# Configuration pour un affichage propre
pd.set_option('display.max_columns', 20)
plt.style.use('ggplot')

# 1. Chargement des UTILISATEURS
u_cols = ['user_id', 'age', 'gender', 'occupation', 'zip_code']
users = pd.read_csv('C:/Users/tchat/Downloads/systeme-de-recommandation/ml-100k/u.user', sep='|', names=u_cols, encoding='latin-1')

# 2. Chargement des FILMS (ITEMS)
# Le fichier contient beaucoup de colonnes (genres), on garde l'essentiel pour l'instant
i_cols = ['movie_id', 'movie_title', 'release_date', 'video_release_date', 'IMDb_URL']
items = pd.read_csv('C:/Users/tchat/Downloads/systeme-de-recommandation/ml-100k/u.item', sep='|', names=i_cols, encoding='latin-1', usecols=[0, 1, 2, 3, 4])

# 3. Chargement des NOTES (RATINGS)
r_cols = ['user_id', 'movie_id', 'rating', 'timestamp']
ratings = pd.read_csv('C:/Users/tchat/Downloads/systeme-de-recommandation/ml-100k/u.data', sep='\t', names=r_cols, encoding='latin-1')

# Affichage des dimensions
print(f"--- Résumé du Dataset ---")
print(f"Utilisateurs : {users.shape[0]}")
print(f"Films        : {items.shape[0]}")
print(f"Notes        : {ratings.shape[0]}")

# Aperçu des notes
print("\n--- Aperçu des notes ---")
print(ratings.head())
```
